```python
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('trainer_state.json', 'rt') as f:
    data = json.load(f)
```

```python
eval_loss = {}
train_loss = {}
for x in data['log_history']:
    if 'loss' in x.keys():
        train_loss[x['epoch']] = x['loss']
    else:
        eval_loss[x['epoch']] = x['eval_loss']
```

```python
plt.figure(figsize=(10, 5))
plt.title('Loss на предобучении модели (MLM)')
plt.plot(train_loss.keys(), train_loss.values(), label='Train loss')
plt.plot(eval_loss.keys(), eval_loss.values(), label='Eval loss')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy')
plt.legend()
plt.savefig('loss_pretrain.pdf')
```
